```python
world_geosn = r'nyc.json' # geojson file
print(world_geosn)
```

```python
world_ny = folium.Map(location=[40.7127837,-74.0059413], zoom_start=11)
#world_ny
```

```python
world_ny.choropleth(
    geo_data=world_geosn,
    data=group_df,
    columns=['neighborhood','count'],
    key_on='feature.properties.neighborhood',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Shooting in NYC'
)

# display map

world_ny
```
